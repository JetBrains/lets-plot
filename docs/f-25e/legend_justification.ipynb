## Legend justification

The `legend_justification` parameter of `theme()` now accepts additional string values: `'top'`, `'right'`, `'bottom'`, and `'left'`. Previously, it only accepted `'center'` as a string or a pair of numeric values, which was not convenient for simple alignment cases.

```python
from lets_plot import *
LetsPlot.setup_html()
```

```python
data = {'x': [0, 0.4, 1], 'y': [0, 0.6, 1], 'Color': ['A', 'B', 'C']}
p = ggplot(data, aes('x', 'y', color='Color')) + geom_point() 
```

```python
# The default legend position and justification
p
```

```python
# Align the legend to the top or bottom
p + theme(legend_justification="top")
```

```python
# When the legend is placed at the top or at the bottom, you can align them to the left or right
p + theme(legend_position="top", legend_justification="left")
```
