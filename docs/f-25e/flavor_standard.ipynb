## Using `flavor_standard()` to return to the theme defaults

Use flavor_standard() to discard other flavor settings and restore the themeâ€™s default colors, without affecting any colors you set manually.

```python
import pandas as pd
from lets_plot import *

LetsPlot.setup_html()
```

```python
mpg_df = pd.read_csv ("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
```

```python
# A base plot
p = ggplot(mpg_df, aes("cty","hwy", color='drv')) + \
    geom_point(tooltips=layer_tooltips().line('@manufacturer @model'))
p
```

```python
customized_plot = p + theme_classic() + flavor_darcula()
customized_plot
```

```python
# Returning back to the theme_classic defaults
customized_plot + flavor_standard()
```

```python
# Note that the effect of flavor_standard() depends on the theme. Here theme_grey() colors are restored.
p + theme_grey() + flavor_darcula() + flavor_standard()
```
