```python
from pandas import DataFrame
import numpy as np
from lets_plot import *

LetsPlot.setup_html()
```

```python
np.random.seed(123)
data = DataFrame(dict(
    cond=np.repeat(['A','B'], 200),
    rating=np.concatenate((np.random.normal(0, 1, 200), np.random.normal(.8, 1, 200)))
))
# Find the mean of each group
cdat = data.groupby(['cond'], as_index=False).mean()

# Create plot and store it variable 'p' 
p = ggplot(data, aes(x='rating')) + geom_histogram(binwidth=.5, colour="black", fill="white") \
+ geom_vline(data=cdat, mapping=aes(xintercept='rating'), linetype="dashed", size=1, colour="red") \
+ facet_grid(None, 'cond') + ggsize(500, 250)
```

### Export SVG to file

```python
from lets_plot.export.simple import export_svg

export_svg(p, "hist.svg")
```

```python
# Load and display saved SVG
from IPython.display import SVG

display(SVG(filename="hist.svg"))
```

### Export HTML to file

```python
from lets_plot.export.simple import export_html

export_html(p, "hist.html")
```

```python
# Load and display saved HTML
from IPython.display import HTML

display(HTML(filename="hist.html"))
```

### Export `iframe` to file

Use `iframe=True` option to wrap plot HTML into `iframe` fitting the plot size.

```python
export_html(p, "hist_iframe.html", iframe=True)
display(HTML(filename="hist_iframe.html"))
```
