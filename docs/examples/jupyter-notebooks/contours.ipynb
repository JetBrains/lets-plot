```python
import numpy as np
from lets_plot import *

LetsPlot.setup_html()
```

```python
# This example was found at: https://jakevdp.github.io/PythonDataScienceHandbook/04.04-density-and-contour-plots.html
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.arange(0, 5, 6./50)
y = np.arange(0, 5, 6./40)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
data = dict(
    x = X.reshape(-1),
    y = Y.reshape(-1),
    z = Z.reshape(-1)
)
```

```python
p = ggplot(data, aes('x', 'y', fill='z'))
# Draw tile plot
p + geom_tile() + scale_fill_hue()
```

```python
# Same but in grey
#  - remove colorbar
p1 = p + geom_tile(alpha=.5) + scale_fill_grey(100, 20, guide='none')
p1
```

```python
# Add contours
p1 + geom_contour(aes(z='z'), color='red')
```

```python
# Set contour color by level
p1 + geom_contour(aes(z='z', color='..level..')) + scale_color_gradient(low='dark_green', high='yellow')
```

### Filling contours by level

```python
p2 = ggplot(data, aes('x', 'y'))

# Filled polygons are not working well in this case
p2 + geom_polygon(aes(z='z',fill='..level..'), stat='contour')
```

```python
# 'contourf' geom works a lot better
p2 + geom_contourf(aes(z='z',fill='..level..'))
```
