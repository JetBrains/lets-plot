```python
from pandas import DataFrame
import numpy as np
from lets_plot import *

LetsPlot.setup_html()
```

```python
np.random.seed(123)
data = DataFrame(dict(
    cond=np.repeat(['A','B'], 200),
    name=np.repeat(['A','B'], 200),
    rating=np.concatenate((np.random.normal(0, 1, 200), np.random.normal(.8, 1, 200)))
))
```

```python
# A basic box plot
p1 = ggplot(data, aes(x='cond', y='rating', fill = 'name')) + ggsize(300, 200) + geom_boxplot()
p1
```

```python
# top-right
p1 + theme(tooltip_anchor='top_right')
```

```python
# top-left
p1 + theme(tooltip_anchor='top_left')
```

```python
# bottom-right
p1 + theme(tooltip_anchor='bottom_right')
```

```python
# bottom-left
p1 + theme(tooltip_anchor='bottom_left')
```

```python
# Load MPG dataset
mpg = pd.read_csv ("https://jetbrains.bintray.com/lets-plot/mpg.csv")
mpg.head ()
```

```python
# Default legend
p2 = ggplot(mpg, aes('displ', 'hwy', color='manufacturer')) + geom_point(size=5) + ggsize(500,250)
p2 + scale_color_discrete(guide=guide_legend(nrow=4)) \
+ theme(legend_position='bottom', tooltip_anchor='top_right') \
+ ggsize(500,400)
```
