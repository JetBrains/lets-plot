```python
from lets_plot.geo_data import *
```

```python
boundaries = regions_country(['Finland', 'Sweden']).boundaries(resolution=4)
boundaries
```

```python
map_theme = theme(axis_line="blank", axis_text="blank", axis_title="blank", axis_ticks="blank")
ggplot() + geom_polygon(aes(fill='found name'), data=boundaries, color="white") \
    + map_theme + scale_fill_hue(name="Country") \
    + coord_map()
```

```python
# For our map let's use raster tiles provided by Wikimedia Foundation.
LetsPlot.set(maptiles_zxy(url='https://a.tile.openstreetmap.org/{z}/{x}/{y}.png'))
ggplot() + geom_livemap() + geom_polygon(aes(fill='found name'), data=boundaries, color="white") \
    + scale_fill_hue(name="Country")
```

# States

```python
boundaries_states_48 = regions_state('us-48').boundaries(resolution=4)
```

```python
states_48_plot = ggplot() \
    + geom_polygon(aes(fill='found name'), data=boundaries_states_48, show_legend=False, color="white") \
    + map_theme \
    + scale_fill_brewer(name="State", type="qual", palette="Dark2") + ggsize(900, 400) + coord_fixed()
states_48_plot
```

```python
alaska_plot = ggplot() + geom_polygon(map=regions_state('alaska').boundaries(4), fill="light-blue", color="#172048") + xlim(-180, -130) + map_theme
hawaii_plot = ggplot() + geom_polygon(map=regions_state('hawaii').boundaries(4), fill="#00A954", color="#D91023") + map_theme

bunch = GGBunch()
bunch.add_plot(states_48_plot, 0, 0)
bunch.add_plot(alaska_plot, 0, 300, 200, 100)
bunch.add_plot(hawaii_plot, 200, 300, 150, 90)
bunch
```

# Builder

```python
warwick_world = regions_builder(request='warwick', level='city') \
    .allow_ambiguous() \
    .build()

ggplot() + geom_map(map=regions_country().boundaries(resolution=1))  \
    + geom_point(map=warwick_world, size=3, color="red") \
    + coord_cartesian(xlim=[-130, 20])
```

```python
boston_us = regions(request='boston', within='us')
warwick_near_boston = regions_builder(request='warwick', level='city') \
    .where('warwick', near=boston_us) \
    .build()
```

```python
ggplot() + geom_map(map=regions_county(within="Massachusetts").boundaries(resolution=7))  \
    + geom_point(map=warwick_near_boston, size=7, color="red") \
    + geom_polygon(map=boston_us)
```

```python
from shapely.geometry import box
warwick_box = regions_builder(request='warwick', level='city') \
    .where('warwick', within=box(-72, 41.5, -71, 42)) \
    .allow_ambiguous() \
    .build()
```

```python
ggplot() + geom_map(map=regions_county(within="Massachusetts").boundaries(resolution=7)) \
    + geom_rect(xmin=-72, ymin=41.5, xmax=-71, ymax=42, alpha=.1) \
    + geom_point(map=warwick_box, size=7, color="red") \
    + geom_polygon(map=boston_us)
```
