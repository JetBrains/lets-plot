```python
from lets_plot import *

LetsPlot.setup_html()
LetsPlot.set_theme(theme_grey())
```

```python
data = {
    'drink': ['coffee','tea','water', 'milk'],
    'mean': [3, 4, 6, 2],
    'upper': [1, 5, 7, 4],
    'lower': [6, 3, 2, 1]
}
```

```python
ggplot(data) + \
    geom_crossbar(aes(x='drink', middle='mean', ymin='lower', ymax='upper', color='mean'),
                  tooltips=layer_tooltips().disable_splitting()) + \
    scale_color_discrete() + theme_grey()
```

```python
ggplot(data) + \
    geom_crossbar(aes(x='drink', middle='mean', ymin='lower', ymax='upper', color='mean'),
                  fill_by='color',
                  tooltips=layer_tooltips().disable_splitting()) + \
    scale_color_discrete()
```

```python
ggplot(data) + geom_crossbar(aes(x='drink', middle='mean', ymin='lower', ymax='upper', color='mean'))
```

```python
ggplot(data) + \
    geom_errorbar(aes(x='drink', y='mean', ymin='lower', ymax='upper', color='mean')) + \
    scale_color_discrete()
```

```python
ggplot(data) + \
    geom_pointrange(aes(x='drink', y='mean', ymin='lower', ymax='upper', fill='mean'), \
                    size=3, linewidth=3, shape=21, color='dark_green') + \
    scale_fill_discrete()
```

```python
ggplot(data) + \
    geom_pointrange(aes(x='drink', y='mean', ymin='lower', ymax='upper', fill='mean'),
                    size=3, linewidth=3, shape=21, color='dark_green', alpha=0.5) + \
    scale_fill_discrete()
```
