```python
from lets_plot import *
from lets_plot.mapping import as_discrete
```

```python
LetsPlot.setup_html()
```

```python
df1 = {
    'x' : ['c', 'c', 'a', 'a', 'd', 'b', 'b', 'a']
}

gggrid([
    ggplot(df1) + geom_bar(aes('x')),
    ggplot(df1) + geom_bar(aes(as_discrete('x', levels=['a','b','c','d']))),
    ggplot(df1) + geom_bar(aes(as_discrete('x', levels=['a','d']))) # missed values will be append
])
```

```python
df2 = {
  'x' : [3, 2, 1, 3, 3, 1, 4]
}

gggrid([
    ggplot(df2) + geom_bar(aes('x', fill = as_discrete('x'))),
    ggplot(df2) + geom_bar(aes('x', fill = as_discrete('x', levels=[4,1,2,3])))
],ncol=1)
```

```python
ggplot(df2) + geom_bar(aes(as_discrete('x',order=1), fill = as_discrete('x', levels=[4,1,2,3])))
```

```python
df3 = {
    'x'   :  [ 'B', 'A', 'C'],
    'fill':  [ '2', '1', '3' ]
}

p3 = ggplot(df3) + \
    geom_bar(aes(x = as_discrete('x', levels=['A','C','B']), 
                 fill=as_discrete('x', levels=['C','A','B']))) 


print(p3.as_dict()['layers'][0]['data_meta']['series_annotations'])

# Use last specified factor_levels
p3
```
