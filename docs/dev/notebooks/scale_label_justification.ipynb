# Verifying hjust and vjust for axis labels

```python
data = {
    "x": ["OXXXXXXX", "OOOOO", "It's lXXX", "XOOOOXXX"],
    "y": [500, 1000, 500, 0],
    "z": ["a", "b", "c", "d"]
}
```

```python
justifications = [1.0, 0.5, 0.0]
angles=[90, 30, 0, -30, 90]
```

```python
def create_plot(position, angle, hjust, vjust):
    if position == 'top' or position == 'bottom':
        theme_element = theme(axis_text_x=element_text(angle=angle, hjust=hjust, vjust=vjust), axis_title="blank") + \
            scale_x_discrete(position = f'{position}')
        x_data = "x"
    else:
        theme_element = theme(axis_text_y=element_text(angle=angle, hjust=hjust, vjust=vjust), axis_title="blank") + \
            scale_y_continuous(position = f'{position}') 
        x_data = "z"
    return (
        ggplot(data) + 
        geom_point(aes(x=x_data, y="y"), size=5) + 
        theme_classic() + 
        ggtitle(f"h{hjust}, v{vjust}") +
        theme_element
    )

        
def generate_all_plots(position, angle):
    return [create_plot(position=position, angle=angle, hjust=hjust, vjust=vjust)
        for hjust in justifications
        for vjust in justifications]


def create_plot_grid(position):
    for angle in angles:
        print('-'*150)
        print(f'ANGLE: {angle}°')
        (gggrid(generate_all_plots(position, angle), ncol=3) + ggsize(1024, 1024)).show()
```

## Checking hjust for y-axis

Reproduced an example from f-25a/axis_label_justification.ipynb - it contained an error.

```python
cdata = pd.DataFrame({
    'category': [
        'Highly Recommended',
        'Moderately Satisfactory',
        'Needs Improvement',
        'Exceeds Expectations',
        'Barely Acceptable Quality'
    ],
    'value': [25, 40, 15, 35, 10]
})

cp = ggplot(cdata, aes(x='value', y='category')) + \
    geom_bar(stat='identity', fill='#FB7') + \
    xlab("Value") + ylab("Category")
```

```python
cp + ggtitle("Default appearance")
```

* Left-aligning long labels creates a consistent starting edge, improving readability, while aligning with natural left-to-right reading patterns.

```python
cp + \
    ggtitle("Corrected with hjust=0") + \
    theme(
        axis_text_y=element_text(angle=0, hjust=0), 
        plot_inset=[0, 0, 0, 15], 
        axis_text_spacing_y=-20
    )
```
