```python
import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html()
LetsPlot.set_theme(flavor_solarized_dark())
```

```python
mpg_df = pd.read_csv("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
mpg_df.head(3)
```

```python
p = ggplot(mpg_df) + geom_pie(aes(fill='class', weight='displ')) + labs(title="Plot title", caption="Plot caption")
p
```

```python
p + theme(plot_margin = margin(t=20, r=10, b=40, l=50))
```

```python
p + theme(plot_margin = margin(t=20, r=10, b=40, l=50)) + ggsize(400,300)
```

```python
p_facet = p + facet_wrap(facets='fl', ncol=2)
p_facet
```

```python
p_facet + theme(plot_margin = margin(t=20, r=10, b=40, l=50))
```

```python
p_facet + theme(plot_margin = margin(t=20, r=10, b=40, l=50)) + ggsize(600,600)
```

```python
gggrid([
    p + theme(plot_margin = margin(t=20, r=20, b=5, l=5)),
    p_facet + theme(plot_margin = margin(t=5, r=5, b=20, l=20))
])
```

```python
# move to the center

p + coord_fixed()
```

```python
# apply plot_margin

p + coord_fixed() + theme(plot_margin = margin(t=20, r=10, b=10, l=140))
```

```python
gggrid([
    ggplot() + geom_livemap() + ggtitle("Default - 5 around"),
    ggplot() + geom_livemap() + ggtitle("Apply plot_margin") + theme(plot_margin = margin(t=10, r=10, b=20, l=20)),
])
```

```python
gggrid([
   p + theme(plot_margin = margin(t=20, b=20)),
   ggplot() + geom_livemap() + ggtitle("Livemap") + theme(plot_margin = margin(t=20, b=20))    
])
```
