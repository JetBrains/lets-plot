```python
import numpy as np
from lets_plot import *

LetsPlot.setup_html()
```

```python
n = 100
np.random.seed(42)
x = np.random.choice(['a', 'b', 'b', 'c'], size=n)
y = np.random.normal(size=n)

p = ggplot({'x': x, 'y': y}, aes(x='x', y='y'))
```

```python
box_plot = p + geom_boxplot(fatten=5, varwidth=True, outlier_shape=8, outlier_size=5)

gggrid(
    plots = [
        box_plot,
        box_plot + flavor_darcula(),
        box_plot + flavor_solarized_light(),
        box_plot + flavor_solarized_dark(),
        box_plot + flavor_high_contrast_light(),
        box_plot + flavor_high_contrast_dark()
    ],
    ncol = 2
)
```

```python
points = p + geom_jitter(aes(size='y'), tooltips=layer_tooltips().line('Tooltip'))

gggrid(
    plots = [
        points,
        points + flavor_darcula(),
        points + flavor_solarized_light(),
        points + flavor_solarized_dark(),
        points + flavor_high_contrast_light(),
        points + flavor_high_contrast_dark()
    ],
    ncol = 2
)
```

```python
# Filled circle (shape=21) looks better
filled_points = p + geom_jitter(aes(size='y'), shape=21, tooltips=layer_tooltips().line('Tooltip'))

gggrid(
    plots = [
        filled_points,
        filled_points + flavor_darcula(),
        filled_points + flavor_solarized_light(),
        filled_points + flavor_solarized_dark(),
        filled_points + flavor_high_contrast_light(),
        filled_points + flavor_high_contrast_dark()
    ],
    ncol = 2
)
```

```python
bar = p + geom_bar()

gggrid(
    plots = [
        bar,
        bar + flavor_darcula(),
        bar + flavor_solarized_light(),
        bar + flavor_solarized_dark(),
        bar + flavor_high_contrast_light(),
        bar + flavor_high_contrast_dark()
    ],
    ncol = 2
)
```
