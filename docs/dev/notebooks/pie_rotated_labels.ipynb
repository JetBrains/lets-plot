```python
import pandas as pd

from lets_plot import *
from lets_plot.mapping import *
```

```python
LetsPlot.setup_html() 
```

```python
LetsPlot.set_theme(theme_void())
```

```python
mpg_df = pd.read_csv("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
mpg_df.head(3)
```

```python
ggplot(mpg_df) + \
    geom_pie(aes(fill="class"), size=18,
             labels=layer_labels()
                          .line('@..proppct.. %')
                          .format("@..proppct..", ".1f")
                          .size(18))
```

```python
ggplot(mpg_df) + \
    geom_pie(aes(fill=as_discrete("manufacturer", order_by="..count..", order=-1)), 
             size=20, hole=0.2,
             labels=layer_labels()
                         .line("@..proppct.. %")
                         .format("@..proppct..", ".1f")) + \
    guides(fill=guide_legend(ncol=3)) + \
    theme(legend_position="bottom")
```

```python
length = {
  'name' : ['20-50 km', '50-75 km', '10-20 km', '75-100 km', '3-5 km', '7-10 km', '5-7 km', '>100 km', '2-3 km'],
  'count': [1109, 696, 353, 192, 168, 86, 74, 65, 53],
  'explode': [0, 0, 0, 0.1, 0.1, 0.2, 0.3, 0.4, 0.6]
}

ggplot(length) + \
    geom_pie(aes(fill='name', slice='count', explode='explode'), stat='identity',
             size=20, hole=0.2,
             labels=layer_labels(['name']).size(11)) + \
    scale_fill_gradient(low='dark_blue', high='light_green')
```
