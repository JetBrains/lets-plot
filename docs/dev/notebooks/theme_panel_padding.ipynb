```python
from lets_plot import *
LetsPlot.setup_html()

from lets_plot.plot.coord import coord_polar
```

```python
def dump_spec(plot, display=None):
    import json

    try:
        import clipboard
    except:
        clipboard = None
        
    from lets_plot._type_utils import standardize_dict
    
    plot_dict = standardize_dict(plot.as_dict())
    plot_json = json.dumps(plot_dict, indent=2)
    
    if clipboard:
        clipboard.copy('')
        clipboard.copy(str(plot_json))
    else:
        if display is None:
            display = True
    
    if display:
        print(plot_json)

    return plot
```

```python
p = ggplot() \
    + geom_point(aes(x=[1, 2, 3, 4, 5, 6])) \
    + theme_grey()
```

# Default

```python
gggrid([
    p,
    p + coord_polar()
])
```

# Padding=40

```python
padding = 40
dump_spec(gggrid([
    p + theme(panel_padding=padding) + ggtitle('padding=' + str(padding)),
    p + theme(panel_padding=padding) + coord_polar() + ggtitle('padding=' + str(padding)),
], ncol=2))
```

```python
p \
    + theme(panel_padding=padding) \
    + coord_fixed() \
    + ggtitle('coord_fixed() + padding=' + str(padding))
```
