# Scale Break Filtering

This notebook checks that axis labels stay paired with their computed offsets when scale breaks are filtered before rendering. The corr-plot examples stress dense square-axis labels (including rotated ±90° labels on the vertical axis), while the polar example includes out-of-range angular breaks and compares different `hjust`/`vjust` settings.

```python
import pandas as pd
from lets_plot import *
from lets_plot.bistro.corr import *

LetsPlot.setup_html()
```

```python
housing_df = pd.read_csv("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/Ames_house_prices_train.csv")
housing_df.shape
```

```python
housing_corr = housing_df
```

```python
p = corr_plot(housing_corr, threshold=.5).tiles(diag=False).palette_BrBG().build()

gggrid([p + theme(axis_text_x=element_text(angle=90)),
    p + theme(axis_text_x=element_text(angle=90, vjust=0), axis_text_y=element_text(hjust=0)),
    p + theme(axis_text_x=element_text(angle=90, vjust=0.5), axis_text_y=element_text(hjust=0.5)) 
   ]) + ggsize(1200, 400)
```

## Vertical axis with rotated (±90°) labels

Rotated y-axis labels are filtered for overlap using each label's **true rendered box** (text anchor + offset), not a raw bounding-box estimate. For ±90° text a label's extent *along* the axis equals its (variable) text width, so a box that does not match the rendered position makes the filter drop the wrong labels. `vjust` slides each label along the axis — for every `vjust` the kept labels should stay paired with their ticks and must not visually overlap.

```python
gggrid([
    p + theme(axis_text_y=element_text(angle=90)) + ggtitle('y-axis angle=90 (default vjust)'),
    p + theme(axis_text_y=element_text(angle=90, vjust=0)) + ggtitle('vjust=0 (labels lower)'),
    p + theme(axis_text_y=element_text(angle=90, vjust=1)) + ggtitle('vjust=1 (labels higher)'),
]) + ggsize(1200, 400)
```

```python
polar_df = pd.DataFrame({
    'angle': list(range(24)),
    'value': [5 + (i * 7) % 10 for i in range(24)]
})

polar_base = (ggplot(polar_df, aes('angle', 'value'))
    + geom_path()
    + geom_point(size=3)
    + coord_polar(xlim=[0, 24], ylim=[0, 16])
)

polar_base
```

```python
polar_breaks = [-6, 0, 3, 6, 9, 12, 15, 18, 21, 24, 30]
polar_labels = [f'{b}:00' for b in polar_breaks]

polar_p = polar_base + scale_x_continuous(breaks=polar_breaks, labels=polar_labels)

gggrid([
    polar_p + ggtitle('polar filtered breaks'),
    polar_p + theme(axis_text_x=element_text(hjust=0, vjust=0)) + ggtitle('hjust/vjust = 0'),
    polar_p + theme(axis_text_x=element_text(hjust=0.5, vjust=0.5)) + ggtitle('hjust/vjust = 0.5')
]) + ggsize(1100, 350)
```
