# Parameter `trim` for `geom_histogram()`
The `trim` parameter controls how the stat is computed. With `trim=False` (the default value), the stat is computed on the full range of data. With `trim=True`, the stat is computed over the range of a group or facet. This enables the usage of the `scales='free_x'` parameter in the `facet_wrap()` and `facet_grid()` functions.

```python
from lets_plot import *
from lets_plot.mapping import as_discrete

LetsPlot.setup_html()
```

```python
d = {
    'v': [1, 2, 3, 10, 11, 12],
    'c': [1, 1, 1, 2, 2, 2]
}

p = ggplot(d, aes(x='v', fill=as_discrete('c'))) + facet_wrap(facets='c', scales='free_x')
```

With `trim=False` the `scales='free_x'` doesn't affect the x-axis:

```python
p + geom_histogram()
```

`trim=True` fixes the `scales='free_x'` parameter:

```python
p + geom_histogram(trim=True)
```
