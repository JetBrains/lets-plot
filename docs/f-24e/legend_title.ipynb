## Legend title
The legend title can be specified in guide_legend() and guide_colorbar().

```python
import pandas as pd
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
mpg = pd.read_csv ("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
mpg.head(3)
```

```python
# Legend title

p = ggplot(mpg) + geom_point(aes('displ', 'hwy', color='cty', shape='cyl'), size=4) 
p + guides(shape=guide_legend(title="Cylinders", ncol=2), \
    color=guide_colorbar(title="City mileage"))
```
