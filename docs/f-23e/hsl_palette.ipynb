# HSL color model.

Lets-Plot now uses the `HSL` color model instead of `HLV`. This model is closer to a model, used by `ggplot2`, and often used in other libraries. The most notable change is a way how the `l` parameters affects the resulting color. 


```python
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data = { 'x': ['A', 'B', 'C', 'D', 'E', 'F'] }
p = ggplot(data) \
    + geom_boxplot(aes(x='x', fill='x'), stat='identity', lower=25, middle=50, upper=75, ymin=0, ymax=100)
```

Default discrete palette:

```python
p + scale_fill_hue()
```

Default continuous palette:

```python
data = {
    'x': list(range(60))
}

ggplot(data) \
    + geom_tile(aes(x='x', fill='x')) \
    + scale_fill_hue() \
    + coord_cartesian() \
    + ggsize(600, 200)
```

Parameter `l` now correctly controls lightness:

```python
p + scale_fill_hue(c=95, l=75) + ggtitle('scale_fill_hue(c=95, l=75)')
```

Palette with the same parameters, produced by **previous** versions of lets-plot:

```python
from IPython.display import SVG, Image, HTML
display(Image('hsl_palette/luminance_incorrect_handling.png'))
```

`h_start` parameter now works with descrete data:

```python
p + scale_fill_hue(h_start=180)
```
