# Annotations for bar plot


Annotations are configured using the `labels` parameter.


```python
import pandas as pd

from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
mpg_df = pd.read_csv ("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
mpg_df.head(4)
```

#### `position='stack'`

```python
ggplot(mpg_df, aes('class', fill='drv')) + \
    theme(legend_position='none') + \
    geom_bar(labels=layer_labels().line('@..proppct..').format('..proppct..', '{d}%'),
             tooltips='none')
```

#### `position='dodge'`

```python
ggplot(mpg_df, aes('class', fill='drv')) + \
    theme(legend_position='none') + \
    geom_bar(labels=layer_labels().line('@..proppct..').format('..proppct..', '{d}%'),
             tooltips='none', position='dodge')
```

#### `position='fill'`

```python
ggplot(mpg_df, aes('class', fill='drv')) + \
    coord_flip() + \
    theme(legend_position='none') + \
    geom_bar(labels=layer_labels().line('@..proppct..').format('..proppct..', '{d}%'), 
             tooltips='none', 
             position='fill')
```
