```python
import numpy as np
import rasterio as ro

from lets_plot import *
LetsPlot.setup_html()
```

```python
# Load Mount Rainier shaded relief 
# Source: https://pugetsoundlidar.ess.washington.edu/lidardata/restricted/filegeodatabase/mtrainierNP/index.html

data = ro.open('./data/mount_rainier_shaded_relief.tif')

crs=data.crs
bounds=data.bounds

print("{}\n{}".format(crs, bounds))
```

```python
# Get the first band as a 2D numpy array
image = data.read(1)
type(image), np.shape(image)
```

```python
extent = [bounds.left, bounds.right, bounds.bottom, bounds.top]

ggplot() + geom_imshow(image, extent=extent) + theme_void()
```

```python
# Custom colormap

high_blue = '#97C3DC'
low_blue = '#1B63A5'

ggplot() + geom_imshow(image, cmap=[low_blue, high_blue], extent=extent) + theme_void()
```

```python
# Use generated colorc.

palette = (
    scale_color_gradient2(
        low=low_blue, 
        mid='white', 
        high=high_blue, 
        midpoint=0.5)       # <-- the midpoint must be in range -1..1 
    .palette(30)            # <-- generate 30 colors
)

ggplot() + geom_imshow(image, cmap=palette, extent=extent) + theme_void()
```
