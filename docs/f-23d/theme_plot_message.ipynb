### Option to show/hide plot messages
This option allows to hide messages produces by sampling.

```python
from lets_plot import *
LetsPlot.setup_html()

d = { 
    'x': [it for it in range(100)] 
}
```

```python
ggplot(d) + geom_point(aes(x='x'), sampling=sampling_random(n=10)) + theme(plot_message='blank')
```

#### Hiding plot messages for the whole notebook 
Using the `LestPlot.set_theme(...)` function messages can be suppresed for the whole notebook.

```python
LetsPlot.set_theme(theme(plot_message='blank'))
ggplot(d) + geom_point(aes(x='x'), sampling=sampling_random(n=10))
```
