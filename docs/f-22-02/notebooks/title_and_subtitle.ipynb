```python
import pandas as pd

from lets_plot import *
LetsPlot.setup_html()
```

```python
mpg = pd.read_csv ("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
mpg
```

```python
p = ggplot(mpg) + geom_point(aes('displ', 'hwy', color = 'drv'))
p
```

```python
p + ggtitle("The plot title using 'ggtitle()'")
```

```python
p + labs(title = "The plot title using 'labs()'")
```

```python
# Add subtitle using ggtitle()

p + ggtitle("The plot title", subtitle = "The plot subtitle")
```

```python
#  Add subtitle using labs()

p_with_titles = p + labs(title = "The plot title", subtitle = "The plot subtitle")
p_with_titles
```

```python
# Add color for title/title

# 'title' applies to both the plot's title and subtitle
p_with_titles + theme(title=element_text(color='blue'))
```

```python
# 'plot_title' will also apply to the subtitle
p_with_titles + theme(plot_title=element_text(color='blue'))
```

```python
# Set own color for the subtitle
p_with_titles + theme(plot_title=element_text(color='blue'), plot_subtitle=element_text(color='red'))
```

```python
# Multiple lines in title - using `\n`

p + ggtitle(
        "The plot title:\nFuel efficiency for most popular models of car", 
        subtitle = "The plot subtitle:\nPoints are colored by the type of drive train"
    ) + theme(plot_subtitle=element_text(color='gray'))
```

```python
# Tooltips: split a tooltip line using '\n'

ggplot(mpg) + geom_point(aes('displ', 'hwy', color = 'drv'),
                   tooltips=layer_tooltips()
                          .line('@manufacturer\n @model')
                          .line('@|@class')
                          .line('@|@year'))
```
