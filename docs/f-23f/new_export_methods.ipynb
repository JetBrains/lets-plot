# Export Methods of `ggplot()` and `gggrid()`

Use methods `to_svg()`, `to_html()`,`to_png()`,`to_pdf()` to export plots on disc or in file-like objects.



```python
import numpy as np
import io
import os
from IPython import display
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
x1 = np.random.randint(10, size=100)
p1 = ggplot({'x': x1}, aes(x='x')) + geom_bar()

n = 100
x2 = np.arange(n)
y2 = np.random.normal(size=n)
w, h = 200, 150
```

#### Export to a file

```python
p1.to_svg('D:/new_folder/plot.svg')
```

#### Export gggrid() to a file

```python
p2 = ggplot({'x': x2, 'y': y2}, aes(x='x', y='y')) + ggsize(w, h)
gggrid([p2 + geom_point(), p2 + geom_line()]).to_png('grid.png')
```

#### Export to a file-like object
When `path` is a file-like object, the data is exported to it by calling its write() method.
Below is how you can write SVG data into a file-like object without saving it to a file, and use it for display or anything else.

```python
file_like = io.BytesIO()
p1.to_svg(file_like)
display.SVG(file_like.getvalue())
```

#### Export binary data in PNG or PDF format to a file-like object

```python
file_like = io.BytesIO()
p1.to_png(file_like, scale = 1.0)
display.Image(file_like.getvalue())
```

#### Export `gggrid()` to a file-like object

```python
file_like = io.BytesIO()
gggrid([p2 + geom_point(), p2 + geom_line()]).to_svg(file_like)
display.SVG(file_like.getvalue())
```
